# Paso 1: Carga de datos

```python
# Cargamos los datasets proporcionados en una lista
archivos = os.listdir(mi_ruta)
```

```python
dataFrames = {}                                     #diccionario que contendra los datasets de cada mes

for archivo in archivos:                            #recorremos los archivos de la lista
  if archivo.endswith('.csv'):                      #si lo que hay es un .csv
    clave = archivo.split('_')[3].split('.')[0]     #me quedo con el nombre del mes, sera usado como la clave del diccionario
    ruta_completa = os.path.join(mi_ruta , archivo) #obtenemos la ruta completa para leer los datasets
    df = pd.read_csv(ruta_completa)                 #leemos los datasets
    df_ventas = df.copy()                           #creamos una copia de los datasets

    dataFrames[clave] = df_ventas                   #guardamos la copia del dataset en el diccionario usando como clave el nombre del mes
```

```python
#revisamos que los datasets esten cargados
#visualizar el head del dataset de Abril(a modo de ejemplo)
dataFrames['Abril'].head()
```

```python
#Revisamos si hay valores NaN en todos los dataframes del diccionario.
#Devuelve True si hay valores NaN en los dataframes

listaNaN = []
for clave in dataFrames:
  respuesta = dataFrames[clave].isnull().values.any()
  listaNaN.append((clave,respuesta))

for respuesta in listaNaN:
  print(respuesta)
```

#Paso 2: Limpieza de los datos

```python
for clave in dataFrames:
  #Seleccionar un dataframe de diccionario
  dataSelect = dataFrames[clave]

  #Los valores no numericos de Cantidad Pedida y Precio Unitario se vuelven Nan
  dataSelect['Cantidad Pedida'] = pd.to_numeric(dataSelect['Cantidad Pedida'], errors='coerce')
  dataSelect['Precio Unitario'] = pd.to_numeric(dataSelect['Precio Unitario'], errors='coerce')

  # Elimino todos los Nan y me quedo solo con los valores numericos
  dataSelect = dataSelect.dropna()

  # Transformo todos los valores numericos al tipo de dato int
  dataSelect['Cantidad Pedida'] = dataSelect['Cantidad Pedida'].astype(int)
  dataSelect['Precio Unitario'] = dataSelect['Precio Unitario'].astype(float)

  #Reescribir el dataframe limpiado
  dataFrames[clave] = dataSelect
```

```python
#visualizar el head del dataset de Abril(a modo de ejemplo)
dataFrames['Abril'].head()
```

```python
#Revisamos si hay valores NaN en todos los dataframes del diccionario.
#Devuelve True si hay valores NaN en los dataframes

listaNaN = []
for clave in dataFrames:
  respuesta = dataFrames[clave].isnull().values.any()
  listaNaN.append((clave,respuesta))

for respuesta in listaNaN:
  print(respuesta)
```

#Paso 3: Preparación de los datos

```python
for clave in dataFrames:
  #Seleccionar un dataframe de diccionario
  dataSelect = dataFrames[clave]

  #Para establecer las columnas de Dia(Numero), Dia(Nombre) y Horas
  dataSelect['Fecha de Compra'] = pd.to_datetime(dataSelect['Fecha de Pedido'])
  dataSelect['Hora'] = dataSelect['Fecha de Compra'].dt.hour
  dataSelect['Dias'] = dataSelect['Fecha de Compra'].dt.day
  dataSelect['Dia-'] = dataSelect['Fecha de Compra'].dt.day_name()

  #Para establecer las columnas de Codigo Postal, Calle y Ciudad
  direccion_df = dataSelect['Dirección de Envio'].str.split(', ',expand=True)
  direccion_df.columns = ['Calle','Ciudad','PO BOX']
  direccion_df[['Estado','Cod Postal']] = direccion_df['PO BOX'].str.split(' ',expand=True)
  dataSelect = pd.concat([dataSelect,direccion_df],axis=True)

  #Reescribir el dataframe con las nuevas columnas
  dataFrames[clave] = dataSelect
```

#Paso 4: Analisis de los datos

```python
#Crear un Diccionario para cada analisis que se vaya a realizar

frameMes = {}       #contendra los dataframes necesarios para responder la preguta 1
frameHora = {}      #contendra los dataframes necesarios para responder la preguta 2
frameCiudad = {}    #contendra los dataframes necesarios para responder la preguta 3
frameEstado = {}    #contendra los dataframes necesarios para responder la preguta 3
frameProducto = {}  #contendra los dataframes necesarios para responder la preguta 4
frameDias = {}      #contendra los dataframes necesarios para responder la preguta 5
frameDia = {}       #contendra los dataframes necesarios para responder la preguta 5


#Tambien creo una lista con los meses, dias , hora y ciudades que se usaran en los dataFrames
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
dias = list(range(1,32))  #de 0 a 31
dia = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
horas = list(range(24))   #de 0 a 23
ciudades = ['Atlanta','Austin','Boston','Dallas','Los Angeles','New York City','Portland','San Francisco','Seattle']
```

##Comportamiento de las ventas, Producto mas vendido:

```python
for clave in dataFrames:
  #Seleccionar un dataframe de diccionario
  dataSelect = dataFrames[clave]

  #Agrupar produtos con cantidad de productos pedida
  Productos_vendidos = dataSelect.groupby('Producto')['Cantidad Pedida'].sum()
  Productos_precios  = dataSelect.groupby('Producto')['Precio Unitario'].mean()
  Productos_vendidos = Productos_vendidos.reset_index()
  Productos_precios = Productos_precios.reset_index()

  # Ordenar de mayor a menor
  Productos_vendidos_ordenados = Productos_vendidos.sort_values(by=['Producto'])
  Productos_precios_ordenados =  Productos_precios.sort_values(by=['Producto'])
  # Renombro la columna 'Producto' de Productos_precios_ordenados para posteriormente sacarlo del dataframe
  Productos_precios_ordenados = Productos_precios_ordenados.rename(columns={'Producto':'Produ'})

  # Uno los dataframes con los valores de Productos_vendidos_ordenados y Productos_precios_ordenados
  dataSelect = pd.concat([Productos_vendidos_ordenados,Productos_precios_ordenados],axis=True)
  # Retiro la columna 'Produ' que anteriormente fue renombrada
  dataSelect = dataSelect.drop(['Produ'],axis=1)
  # Ordeno el dataframe por 'Cantidad Pedida'
  dataSelect = dataSelect.sort_values(by=['Cantidad Pedida'], ascending=False)
  # Le reinicio el index
  dataSelect = dataSelect.reset_index(drop=True)
  # Creo la columna 'Ingresos Totales' con el resultado de multiplicar la cantidad de producto y el precio unitario
  dataSelect['Ingresos Totales'] = dataSelect['Cantidad Pedida'] * dataSelect['Precio Unitario']

  #Reescribir el dataframe con las nuevas columnas
  frameProducto[clave] = dataSelect
```

```python
for clave in dataFrames:
  #Seleccionar un dataframe de diccionario
  dataSelect = dataFrames[clave]

  #Agrupar produtos con cantidad de ventas por hora
  Productos_vendidos = dataSelect.groupby('Hora')['Cantidad Pedida'].count()

  # Ordenar de menor a mayor
  Productos_vendidos_ordenados = Productos_vendidos.sort_values(ascending = True)

  dataSelect = pd.DataFrame({'Hora':Productos_vendidos_ordenados.index, 'Cantidad Pedida': Productos_vendidos_ordenados.values})

  # Reordenar por Horas
  dataSelect = dataSelect.sort_values(by='Hora',ascending = True)
  # Reinicio el index
  dataSelect = dataSelect.reset_index(drop=True)

  #Reescribir el dataframe con las nuevas columnas
  frameHora[clave] = dataSelect
```

##1. Comportamiento de las ventas en los distintos meses:

Para responder a la pregunta 1 del informe se realizaron tres gráficos con Plotly, dos gráficos de barras para mostrar la cantidad de ventas y las ganancias y un gráfico de tarta para mostrar el porcentaje de ventas en los distintos meses.

###● ¿Cómo variaron las ventas a lo largo de los diferentes meses?

```python
totalVentas = []
nombreMeses = []

for mes in meses:
  dataFrameMes = frameProducto[mes]

  totalVentasMes = dataFrameMes['Cantidad Pedida'].sum()
  totalVentas.append(totalVentasMes)
  nombreMeses.append(mes)

#Creo el objeto figura
figuraMes1 = go.Figure()

#Agrego grafico en mi objeto figura
figuraMes1.add_trace(go.Bar(
    x = nombreMeses,
    y = totalVentas,
    name = 'Unidades Vendidas',
    marker = dict(color = totalVentas, colorscale = "RdYlGn", colorbar=dict(title="Total de Ventas")),
    text = totalVentas,
    textposition = 'auto'
    ))

figuraMes1.update_layout(
    title = 'Comportamiento de Ventas por Mes',
    xaxis = dict(title='Meses'),
    yaxis = dict(title='Ventas')
)
```

Como se observa en el grafico se presentó una gran variación de las ventas en los distintos meses siendo Diciembre el mes que más destaca en término de ventas, presumiblemente debido a las fiestas de Navidad y Año Nuevo, que es donde por lo general más se vende.

###● ¿Hubo algún mes que se destacó en términos de ventas?

```python
figuraMes2 = px.pie(frameMeses, values='Unidades Vendidas',
             names='Mes',
             title='Comportamiento de Ventas por Mes - Porcentaje')
figuraMes2.show()
```

Como podemos observar el mes de Diciembre es el que más destaca, con un 13.4% de las ventas, seguido por el mes de Octubre con el 10.9% que posiblemente se deba a las festividades de Halloween.

Para responder estas preguntas se realizaron dos gráficos en Plotly para analizar las horas de actividad y cómo varían esas horas en los distintos meses.

###● ¿Cuál es el momento ideal del día para presentar la publicidad y aumentar la probabilidad de compra?

```python
#Creo el objeto figura
figuraHora1 = go.Figure()

for mes in meses:
  dataFrameMes = frameHora[mes]
  ventas_hora = dataFrameMes.groupby("Hora")["Cantidad Pedida"].sum()


  figuraHora1.add_trace(go.Scatter(
      x = horas,
      y = ventas_hora,
      name = mes,
      line = dict(width= 2, shape= 'spline')
      ))


figuraHora1.update_layout(
    title = 'Unidades Vendidas por Hora',
    xaxis = dict(title='Hora', tickvals = horas),
    yaxis = dict(title='Unidades Vendidas' , side="left"),
    )
```

###● ¿Cómo cambian los patrones de ventas por hora a lo largo del año?

```python
figuraHora3 = go.Figure()

horas_especificas = [11,12,13,18,19,20]  #los dos picos mas notorios de figuraHora1

for hora in horas_especificas:
  total_productos_vendidos = []


  for mes in meses:
    dataFrameMes = frameHora[mes]

    ventas_hora = dataFrameMes[dataFrameMes['Hora'] == hora]

    total_productos_vendidos.append(ventas_hora['Cantidad Pedida'].sum())


  figuraHora3.add_trace(go.Scatter(

      x=meses,
      y=total_productos_vendidos,
      mode='lines',
      name=f'{hora} horas',
      line = dict(width= 2, shape= 'spline')
  ))


# Configuro los ejes y los títulos
figuraHora3.update_layout(

  title="Unidades vendidas por hora (Especificas)",
  xaxis=dict(title="Hora", tickvals=horas),  # Para que en el eje x se muestren todos los valores uso tickvals=Horas
  yaxis=dict(title="Unidades vendidas", side="left"))


figuraHora3.show()
```

Podemos observar el cambio en los patrones de ventas por hora a lo largo del año,  se observa que durante Abril, Octubre y Diciembre las horas pico en ventas presentan un aumento considerable, pasando de las 600 unidades vendidas en Enero a las más de 1800 unidades en el mes de Diciembre.

Como puede observarse los productos más vendidos en general y en cada mes fueron las pilas AAA y las pilas AA, sin importar qué mes se observe las ventas de pilas sobrepasan por mucho el de cualquier otro producto.
